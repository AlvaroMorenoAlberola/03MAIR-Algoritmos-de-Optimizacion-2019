```
from time import time

# Cálculo del tiempo de ejecución

def calcular_tiempo(f):
  
  def wrapper(*args, **kwargs):
    inicio = time()
    resultado = f(*args, **kwargs)
    tiempo = time() - inicio
    print("Tiempo de ejecución para el algoritmo: " + str(tiempo))
    return resultado
  
  return wrapper
```

```
#quick_sort

A = [9187, 244, 4054, 9222, 8373, 4993, 5265, 5470, 4519, 7182, 2035, 3506, 4337, 7580, 2554, 2824, 8357, 4447, 7379]

def quick_sort(A):
  if len(A) == 1:
    return A
  elif len(A) == 2:
    return [min(A), max(A)]
  
  pivote = (A[0] + A[1] + A[2]) / 3
  
  lista_izq = []
  lista_der = []
  
  for i in A:
    if i <= pivote:
      lista_izq.append(i)
    else:
      lista_der.append(i)
      
  return quick_sort(lista_izq) + quick_sort(lista_der)

@calcular_tiempo
def QS(A):
  return quick_sort(A)

print(QS(A))
```

```
# Algoritmo voraz

sistema_monetario = [25, 10, 5, 1]

@calcular_tiempo
def cambio_monedas(cantidad, sistema):
  solucion = [0 for i in range(len(sistema))]
  valor_acumulado = 0
  
  for i in range(len(sistema)):
    monedas = int((cantidad - valor_acumulado) / sistema[i])
    solucion[i] = monedas
    valor_acumulado += monedas*sistema[i]
    
    if cantidad == valor_acumulado:
      return solucion
  

cambio_monedas(101, sistema_monetario)
```

```
# Backtracking

n = 4
solucion = [0 for i in range(n)]
etapa = 0

def es_prometedora(solucion, etapa):
  for i in range(etapa+1):
    if solucion.count(solucion[i]) > 1:
      return False
    
  for j in range(i+1, etapa+1):
    if abs(i-j) == abs(solucion[i]-solucion[j]):
      return False
    
  return True

def escribe(sol):
  n = len(sol)
  for x in range(n):
    print("")
    for i in range(n):
      if solucion[i] == x+1:
        print(" X ", end="")
      else:
        print(" · ", end="")
  print("")
  print("")

# n = número de reinas
# solucion = momento en el que estamos de la solución
# etapa = etapa en la que estamos (puede extraerse de la solución, pero pasarla como parámetro ahorrará cálculos)
def reinas(n, solucion, etapa):
  
  for i in range(1, n+1):
    solucion[etapa] = i
  
    if es_prometedora(solucion, etapa):
      if etapa == n-1:
        print(solucion)
        escribe(solucion)
      else:
        reinas(n, solucion, etapa+1)
    else:
      None

    solucion[etapa] = 0

reinas(n, solucion, etapa)
```
